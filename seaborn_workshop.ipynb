```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
```

```python
#1.scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time",size="size",palette="deep")    
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show
```

```python
#2.line plot
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex",ci=None, marker="o")
plt.title("Line Plot of Total Bill vs Tip")
```

```python
#3.bar plot
sns.barplot(data=tips, x="day", y="total_bill", hue='sex', palette="muted")
plt.title("Bar Plot of Total Bill by day")
plt.show()
```

```python
#4.box plot
sns.boxplot(data=tips, x="day", y="tip",hue='smoker', palette="Set2")
plt.title("Box Plot of Tips by Day and Smoker Status")
plt.show()
```

```python
#5.violin plot
sns.violinplot(data=tips, x="day", y="total_bill", hue="time", split=True, palette="Pastel1")
plt.title("Violin Plot of Total Bill by Day and Time")
plt.show()
```

```python
#6.count plot
sns.countplot(data=tips, x="day", hue="smoker", palette="dark")
plt.title("Count Plot of Days with Smoker Status")  
plt.show()
```

```python
#7.regression plot
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

```python
#8.histogram Plot
sns.histplot(data=tips, x="total_bill", bins=20, kde=True,color="blue")
plt.title("Histogram of Total Bill with KDE")   
plt.show()
```

```python
#9.pairplot
sns.pairplot(tips, hue='sex',vars=["total_bill", "tip", "size"], palette="husl")
plt.suptitle("Pairplot :numerical variables by gender", y=1.02)
plt.show()
```

```python
#10.category plot
sns.catplot(data=tips, x="day", y="tip",hue='sex',kind='point',palette='bright')
plt.title("Category Plot of Tips by day and gender")
plt.show()
```

```python
#11.joint plot
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter",hue="smoker", palette="coolwarm")
plt.suptitle("Joint Plot of Total Bill and Tip by Smoker Status", y=1.02)
plt.show()
```

```python
#12.facet grid      
g =sns.FacetGrid(tips, col="time", row="smoker", margin_titles=True).map(sns.histplot, "total_bill", kde=True, color="purple"  )
```

```python
#13.strip plot
sns.stripplot(data=tips, x="day", y="total_bill", hue='sex', jitter=True, dodge=True, palette="Set1")
plt.title("Strip Plot o:tips by data and gender")
plt.show()
```

```python
#14.kde plot
sns.kdeplot(data=tips, x="total_bill", hue='sex', fill=True, palette="tab10")
plt.title("KDE Plot :total bill density by gender")
plt.show()
```
