```python
import pandas as pd
data={
     "Month":['Jan','Feb','Mar','Apr','May','Jun'],
    "Sales":[10000,12000,15000,13000,17000,16000],
    "Profit" :[2000,3000,4000,2500,3500,3000]
}
df=pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.plot(df['Month'],df['Sales'],color='blue',marker ='o',linestyle='-',label='Sales')
plt.title('Sales trends over Months')
plt.xlabel('Month')
plt.ylabel('Sales($)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
width=0.3
plt.bar(df['Month'],df['Sales'],width=width, color='skyblue',label='Sales')
plt.bar(df['Month'],df['Profit'],width=width, color='green',label='Profit',bottom=df['Sales'])
plt.title('Sales and profit trends by  Months')
plt.xlabel('Month')
plt.ylabel('Amount($)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Profit'],labels=df['Month'],autopct='%1.2f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title('profit by month')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Profit'],df['Sales'],color='green',s=100,edgecolors='black')
plt.title('Sales vs Profit scatter graph')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df['Sales'],bins=5,color='purple',edgecolors='black')
plt.title('Sales Distribution ')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.boxplot(df['Profit'],vert=False,patch_artist=True,boxprops=dict(facecolor='lightgreen')),
plt.title('Sales Box plot'),
plt.ylabel('Sales'),
plt.tight_layout(),
plt.show()
```

```python

import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Month":['Jan','Feb','Mar','Apr','May','Jun'],
    "Sales":[10000,12000,15000,13000,17000,16000],
    "Profit" :[2000,3000,4000,2500,3500,3000]
}
df=pd.DataFrame(data)
def generate_plot(plot_type):
    fig=plt.figure(figsize=(8,5))
    if plot_type=="Line plot":
        plt.plot(df['Month'],df['Sales'],color='blue',marker ='o',linestyle='-',label='Sales')
        plt.title('Sales trends over Months')
        plt.xlabel('Month')
        plt.ylabel('Sales($)')
        plt.grid(True)
        plt.legend()
    elif plot_type=="stacked bar chart":
        plt.figure(figsize=(6,3))
        width=0.3
        plt.bar(df['Month'],df['Sales'],width=width, color='skyblue',label='Sales')
        plt.bar(df['Month'],df['Profit'],width=width, color='green',label='Profit',bottom=df['Sales'])
        plt.title('Sales and profit trends by  Months')
        plt.xlabel('Month')
        plt.ylabel('Amount($)')
        plt.grid(True)
        plt.legend()
    elif plot_type=="pie chart":
        plt.figure(figsize=(7,7))
        plt.pie(df['Profit'],labels=df['Month'],autopct='%1.2f%%',startangle=140,colors=plt.cm.Paired.colors)
        plt.title('profit by month')  
    elif plot_type=="scatter plot":
        plt.scatter(df['Profit'],df['Sales'],color='green',s=100,edgecolors='black')
        plt.title('Sales vs Profit scatter graph')
        plt.xlabel('Sales')
        plt.ylabel('Profit')
        plt.tight_layout()
        plt.grid(True)     
    elif plot_type=="histo plot": 
        plt.hist(df['Sales'],bins=5,color='purple',edgecolors='black')
        plt.title('Sales Distribution ')
        plt.xlabel('Sales')
        plt.ylabel('Frequency') 
    elif plot_type=="box plot":       
        plt.boxplot(df['Profit'],vert=False,patch_artist=True,boxprops=dict(facecolor='lightgreen')),
        plt.title('Sales Box plot'),
        plt.ylabel('Sales'),
        plt.tight_layout(),
        return fig
    demo = gr.Interface(fn=generate_plot,inputs=gr.Radio(["Line plot","stacked bar chart","pie chart","scatter plot","histo plot","box plot"],label="choose plot type"),
            outputs=gr.Plot(label="Sales data visualization"),
            title="Sales and Profit visual insight",
            description="choose the type to visualize the data"),
    demo.launch()    
```
